### Cleaning Data

```python
# Mengubah Tipe Data INT menjadi tipe data kategori pata kolom weekday untuk mempermudah analisis data
# Karena berdasarkan pertanyaan bisnis kita hanya ingin mengubah tipe data dari kolom weekday saja

days_df["weekday"] = days_df["weekday"].astype("category")
hours_df["weekday"] = hours_df["weekday"].astype("category")
```

```python
# Mengubah Tipe data Objeck ke Datetime untuk kolom dteday

days_df["dteday"] = pd.to_datetime(days_df["dteday"])
hours_df["dteday"] = pd.to_datetime(hours_df["dteday"])
```

```python
days_df.info()
hours_df.info()
```

```python
# Mengubah Nama Kolom pada kolom weekday agar lebih mudah dipahami dan merepresentasikan nama hari bukan angka

days_df.weekday.replace((0,1,2,3,4,5,6),("Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"),inplace=True)
hours_df.weekday.replace((0,1,2,3,4,5,6),("Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"),inplace=True)
```

```python
# Menggabungkan Kolom dteday dan hr untuk membentuk tanggal dan waktu yang nantinya akan dipakai untuk visualisasi data
hours_df['datetime'] = hours_df.apply(lambda row: row['dteday'] + pd.Timedelta(hours=row['hr']), axis=1)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Perbandingan Antara pemakaian sepeda pada tiap kategori pengguna(casual & registered)

```python
# Membuat figure dan 2 subplot
fig, ax = plt.subplots(2, 1, figsize=(18, 10))

# Plot pengguna casual di subplot pertama
ax[0].plot(days_df['dteday'], days_df['casual'], label='Pengguna Casual', color='b', linewidth=1.5)
ax[0].set_title('Jumlah Pengguna Casual Berdasarkan Waktu', fontsize=14)
ax[0].set_xlabel('Waktu Penyewaan Sepeda', fontsize=12)
ax[0].set_ylabel('Jumlah Penyewaan Sepeda', fontsize=12)
ax[0].legend()
ax[0].grid(True)

# Plot pengguna registered di subplot kedua
ax[1].plot(days_df['dteday'], days_df['registered'], label='Pengguna Registered', color='g', linewidth=1.5)
ax[1].set_title('Jumlah Pengguna Registered Berdasarkan Waktu', fontsize=14)
ax[1].set_xlabel('Waktu Penyewaan Sepeda', fontsize=12)
ax[1].set_ylabel('Jumlah Penyewaan Sepeda', fontsize=12)
ax[1].legend()
ax[1].grid(True)


# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
# Membuat figure dan 2 subplot
fig, ax = plt.subplots(2, 1, figsize=(18, 10))

# Plot pengguna casual di subplot pertama
ax[0].plot(hours_df['datetime'], hours_df['casual'], label='Pengguna Casual', color='b', linewidth=1.5)
ax[0].set_title('Jumlah Pengguna Casual Berdasarkan Waktu', fontsize=14)
ax[0].set_xlabel('Waktu Penyewaan Sepeda', fontsize=12)
ax[0].set_ylabel('Jumlah Penyewaan Sepeda', fontsize=12)
ax[0].legend()
ax[0].grid(True)

# Plot pengguna registered di subplot kedua
ax[1].plot(hours_df['datetime'], hours_df['registered'], label='Pengguna Registered', color='g', linewidth=1.5)
ax[1].set_title('Jumlah Pengguna Registered Berdasarkan Waktu', fontsize=14)
ax[1].set_xlabel('Waktu Penyewaan Sepeda', fontsize=12)
ax[1].set_ylabel('Jumlah Penyewaan Sepeda', fontsize=12)
ax[1].legend()
ax[1].grid(True)


# Menampilkan grafik
plt.tight_layout()
plt.show()
```

### Pertanyaan 2: 
- Bagaimana perbandingan pemakaian sepeda setiap hari dan pada hari apa pemakaian sepeda paling banyak pada data day.csv
- Bagaimana perbandingan pemakaian sepeda setiap hari setiap jam dan pada hari apa dan jam berapa pemakaian sepeda paling banyak pada data hour.csv

```python
# Menghitung total penyewaan sepeda untuk setiap hari
daily_rentals = days_df.groupby('weekday')['cnt'].sum().reset_index()

# Menampilkan hari dengan pemakaian sepeda paling banyak
max_day = daily_rentals.loc[daily_rentals['cnt'].idxmax()]

# Membuat grafik histogram
plt.figure(figsize=(12, 6))
plt.bar(daily_rentals['weekday'], daily_rentals['cnt'], color='skyblue')

# Menambahkan label dan judul
plt.xlabel('Hari', fontsize=12)
plt.ylabel('Jumlah Penyewaan Sepeda', fontsize=12)
plt.title('Pemakaian Sepeda Setiap Hari', fontsize=14)

# Menandai hari dengan pemakaian sepeda paling banyak
plt.annotate(f'Maksimum: {max_day["cnt"]}', 
             xy=(max_day['weekday'], max_day['cnt']), 
             xytext=(max_day['weekday'], max_day['cnt'] + 5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Menampilkan grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
# Mengelompokkan berdasarkan hari (weekday) dan jam (hr), lalu menghitung total penyewaan
grouped = hours_df.groupby(['weekday', 'hr'])['cnt'].sum().reset_index()

# Mengurutkan hasil berdasarkan hari dan jam
grouped_sorted = grouped.sort_values(by=['weekday', 'hr'])

# Membuat grafik bar
plt.figure(figsize=(12, 6))

# Menggunakan pivot untuk membuat format yang sesuai untuk plotting
pivot_table = grouped_sorted.pivot(index='hr', columns='weekday', values='cnt').fillna(0)

# Membuat grafik bar
pivot_table.plot(kind='bar', stacked=False, figsize=(12, 6))

# Menambahkan label dan judul
plt.xlabel('Jam (hr)', fontsize=12)
plt.ylabel('Jumlah Penyewaan Sepeda', fontsize=12)
plt.title('Pemakaian Sepeda Berdasarkan Hari dan Jam', fontsize=14)

# Menampilkan grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Menampilkan legenda
plt.legend(title='Hari', bbox_to_anchor=(1.05, 1), loc='upper left')

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

**Insight:**
- Kita bisa lihat bahwa rentang periode paling banyak terjadi penyewaan sepeda untuk kategori casual adalah 2012-3 sampai 2012-11, sedangkan yang paling sedikit adalah pada awal data diambil yaitu 2011-1 sampai 2011-3
- Sedangkan untuk kategori pengguna registered paling banyak terjadi penyewaan sepeda pada periode 2012-5 sampai 2012-11 sedangkan yang paling sedikit adalah pada periode 2011-1 sampai 2011-5
- Hari jumat merupakan hari dimana pengguna paling banyak penyewa sepeda dan hari minggu merupakan hari paling sedikit orang menyewa sepeda
- Dan pada hari selasa pada jam 17 paling banyak terjadi penyewaan sepeda dan pada jam 0-5 paling sedikit terjadi penyewaan sepeda
